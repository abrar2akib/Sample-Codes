# Wave wise merge

##Wave 3

```python
# Merge B2 with B1 (keep a01 from B2 only)
df_B2B1_3 = df_B2_3.merge(
    df_B1_3.drop(columns=["a01"]),  # drop duplicate a01 from B1
    on="hhid_mid",
    how="left"
)

# Now safe: a01 is still there from B2
df_B2B1A3 = df_B2B1_3.merge(df_A3, on="a01", how="left")

# Merge with Migration module V1 using a01
df_w3 = df_B2B1A3.merge(df_V1_3, on="a01", how="left")
```

## Wave 2

```python
df_B2B1_2 = df_B2_2.merge(
    df_B1_2.drop(columns=["a01"]),  # drop duplicate a01 from B1
    on="hhid_mid",
    how="left"
)

# --- Merge with Module A (household level) ---
df_B2B1A2 = df_B2B1_2.merge(df_A2, on="a01", how="left")

# --- Merge with Migration module V1 (household level) ---
df_w2 = df_B2B1A2.merge(df_V1_2, on="a01", how="left")
```

## Wave 1

```python
# --- Merge B2 (base) with B1 ---
df_B2B1_1 = df_B2_1.merge(
    df_B1_1.drop(columns=["a01"]),  # drop duplicate a01 from B1
    on="hhid_mid",
    how="left"
)

# --- Merge with Module A (household level) ---
df_B2B1A1 = df_B2B1_1.merge(df_A1, on="a01", how="left")

# --- Merge with Migration module V1 (household level) ---
df_w1 = df_B2B1A1.merge(df_V1_1, on="a01", how="left")
```

#Saving

```python
df_w1 = df_w1.convert_dtypes()
df_w2 = df_w2.convert_dtypes()
df_w3 = df_w3.convert_dtypes()
```

```python
# Save as Stata file
df_w1.to_stata("final_2011.dta", write_index=False)
df_w2.to_stata("final_2015.dta", write_index=False)
df_w3.to_stata("final_2018.dta", write_index=False)
```

#Master Merge

```python
df_w1["wave"] = 2011   # 2011
df_w2["wave"] = 2015  # 2015
df_w3["wave"] = 2018   # 2018

df_w1["panel_id"] = df_w1["hhid_mid"]
df_w2["panel_id"] = df_w2["hhid_mid"]
df_w3["panel_id"] = df_w3["hhid_mid"]

df_panel = pd.concat([df_w1, df_w2, df_w3], axis=0, ignore_index=True)
df_panel = df_panel.sort_values(by=["panel_id", "wave"]).reset_index(drop=True)
```

```python
# Get the current columns
cols = df_panel.columns.tolist()

# Reorder: put 'panel_id' and 'wave' first, then all other columns
cols_reordered = ['panel_id', 'wave'] + [c for c in cols if c not in ['panel_id', 'wave']]

# Apply the new order
df_panel = df_panel[cols_reordered]

# Optional: sort by panel_id and wave
df_panel = df_panel.sort_values(by=['panel_id', 'wave']).reset_index(drop=True)
```

```python
df_panel['mig_year'] = df_panel['wave'] - df_panel['v1_03']
df_panel['mig_year'] = df_panel['mig_year'].astype(float)  # convert to integer year
```
